```python
%pylab inline
import pandas as pd
import contributions
import votes
import bills
import sqlCommands
from peoplefinder import PeopleFinder
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import re
from tqdm import tqdm

from operator import attrgetter

def clean_subject(a):
    try:
        #quoteloc = a.index("'") #may get issues with multiple single quotes in one topic.
        #a = a[:quoteloc]+"'"+a[quoteloc:]
        a = re.sub("'","''",a)
    except:
        pass #no quotes found
    #need to add " if there are any spaces in the subject
    #try:
    #    quoteloc = a.index(" ") #will go to except if there are no spaces in the subject
    #    a = '\"'+a+'\"'
    #except:
    #    pass #no spaces found
    return a


#create sqlalchemy engine
dbname = 'legislatr'
engine = sqlCommands.get_engine(dbname)
```

```python
legis = pd.read_sql_table('sponsor_list',engine)
```

```python
names = list()
fnames = list()
lnames = list()
station = list()
state = list()
for bar in legis.sponsor.tolist():
    name = re.sub("\\\\.*\\\\","",bar)
    #name = re.sub("\"","",name)
    #name = ''.join([x for x in name if x != "\""])
    #name = name.replace("\"","")
    name_broken = re.split(",",name)
    station.append(name_broken[0])
    state.append(name_broken[-1])
    fname = re.sub("\"","",name_broken[2])
    lname = re.sub("\"","",name_broken[1])
    fnames.append(fname)
    lnames.append(lname)
    names.append(name_broken[0]+', '+lname+', '+fname)
legis['qsponsor'] = names
legis['fname'] = fnames
legis['lname'] = lnames
legis['station'] = station
legis['state'] = state
```

```python
sqlCommands.write_to_database(dbname,'sponsor_list_det',legis,engine)
```

```python
#Step 1: set up exact name query
#try a name, remember that one name has no first name, so need to be able to deal with that.
totalcount = 0
for i in tqdm(range(0,legis.fname.size)):
    fname = legis.fname.iloc[i]
    lname = legis.lname.iloc[i]
    #print('First Name: ',fname)
    #print('Last Name: ',lname)
    #clean name for query
    fname = clean_subject(fname)
    lname = clean_subject(lname)
    #perfectQ = "SELECT registrantname, amount, honoree, year, registrantid FROM contrib WHERE honoree LIKE '%%"+fname+"%%' AND honoree LIKE '%%"+lname+"%%' AND contributiontype LIKE 'FECA' ;" #definite
    perfectQ = "SELECT * FROM contrib WHERE honoree LIKE '%%"+fname+"%%' AND honoree LIKE '%%"+lname+"%%' AND contributiontype LIKE 'FECA' ;" #definite
    lessQ = "SELECT * FROM contrib WHERE honoree LIKE '%%"+lname+"%%';" #definite + maybe
    perfectdf = pd.read_sql_query(perfectQ,engine)
    lessdf = pd.read_sql_query(lessQ,engine)
    maybedf = pd.concat([perfectdf,lessdf],axis=0).drop_duplicates(keep=False) #maybe
    count = perfectdf['index'].size + maybedf['index'].size
    #print('Count = ',count)
    totalcount = totalcount + count
print('TotalCount = ',totalcount)

```
